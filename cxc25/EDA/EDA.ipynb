```python
import pandas as pd
import numpy as np
```

```python
companies = pd.read_csv('../data/raw_data/companies.csv')
dealInvestor = pd.read_csv('../data/raw_data/dealInvestor.csv')
deals = pd.read_csv('../data/raw_data/deals.csv')
investors = pd.read_csv('../data/raw_data/investors.csv')
ecosystems = pd.read_csv('../data/raw_data/ecosystems.csv')

data = [companies, dealInvestor, deals, investors, ecosystems]
```

making missing data consistent (nullifying)

```python
date_cols = ["dateFounded", "ipoDate", "peDate", "dateAcqusition", "latestRoundDate"]
companies[date_cols] = companies[date_cols].replace("0000-00-00", np.nan)
companies["latestRoundType"] = companies["latestRoundType"].replace("Series ?", np.nan)

print(companies.isnull().sum())
companies.to_csv("../data/EDA/companies_EDA.csv", index=False)
```

```python
missing_round_type = dealInvestor[dealInvestor["roundType"].isnull()]

dealInvestor["roundType"] = dealInvestor["roundType"].replace("Series ?", np.nan)
dealInvestor["date"] = dealInvestor["date"].replace("0000-00-00", np.nan)
dealInvestor["date"] = dealInvestor["date"].fillna(f"{dealInvestor["year"]}-01-01")
dealInvestor["headquarters"] = dealInvestor["headquarters"].fillna(dealInvestor["ecosystemName"])
dealInvestor.drop_duplicates(keep="first")  

print(dealInvestor.isnull().sum())
dealInvestor.to_csv("../data/EDA/dealInvestor_EDA.csv", index=False)
```

```python
deals["roundType"] = deals["roundType"].replace("Series ?", np.nan)
deals["amount"] = deals["amount"].replace("0", np.nan)
deals["date"] = deals["date"].replace("0000-00-00", np.nan)
deals["year"] = deals["year"].replace("0000", np.nan)
deals["headquarters"] = deals["headquarters"].fillna(deals["ecosystemName"])

print(deals.isnull().sum())
deals.to_csv("data/EDA/deals_EDA.csv", index=False)
```

```python
print(investors.isnull().sum())
investors.to_csv("data/EDA/investors_EDA.csv", index=False)
```
